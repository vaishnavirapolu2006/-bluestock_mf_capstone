```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Paths
BASE_DIR = Path().resolve()
PROCESSED_DIR = BASE_DIR / "data" / "processed"

# Load cleaned files
nav_history = pd.read_csv(PROCESSED_DIR / "02_nav_history.csv", parse_dates=["date"])
fund_master = pd.read_csv(PROCESSED_DIR / "01_fund_master.csv")
performance = pd.read_csv(PROCESSED_DIR / "07_scheme_performance.csv")
aum = pd.read_csv(PROCESSED_DIR / "03_aum_by_fund_house.csv")
sip = pd.read_csv(PROCESSED_DIR / "04_monthly_sip_inflows.csv")
benchmark = pd.read_csv(PROCESSED_DIR / "10_benchmark_indices.csv", parse_dates=["date"])

print("✅ All files loaded!")
```

```python
# Chart 1 — NAV growth of all funds over time
plt.figure(figsize=(14, 6))

for code in nav_history["amfi_code"].unique():
    df = nav_history[nav_history["amfi_code"] == code]
    plt.plot(df["date"], df["nav"], label=str(code))

plt.title("NAV Growth of All Funds (2022–2025)")
plt.xlabel("Date")
plt.ylabel("NAV (₹)")
plt.legend(fontsize=7, ncol=2)
plt.tight_layout()
plt.show()
```

```python
# Chart 2 — Top fund houses by AUM
latest_aum = aum[aum["date"] == aum["date"].max()]
latest_aum = latest_aum.sort_values("aum_crore", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=latest_aum, x="aum_crore", y="fund_house", palette="Blues_r")
plt.title("Top 10 Fund Houses by AUM")
plt.xlabel("AUM (₹ Crore)")
plt.ylabel("Fund House")
plt.tight_layout()
plt.show()
```

```python
# Chart 2 — Top fund houses by AUM (fixed)
latest_aum = aum[aum["date"] == aum["date"].max()]
latest_aum = latest_aum.sort_values("aum_crore", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=latest_aum, x="aum_crore", y="fund_house", hue="fund_house", legend=False)
plt.title("Top 10 Fund Houses by AUM")
plt.xlabel("AUM (₹ Crore)")
plt.ylabel("Fund House")
plt.tight_layout()
plt.show()
```

```python
# Chart 3 — SIP inflows over time
plt.figure(figsize=(14, 5))
plt.plot(sip["month"], sip["sip_inflow_crore"], color="green", marker="o", markersize=3)
plt.title("Monthly SIP Inflows Over Time (₹ Crore)")
plt.xlabel("Month")
plt.ylabel("SIP Inflow (₹ Crore)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Chart 4 — Category wise inflows
import pandas as pd
cat_df = pd.read_csv(PROCESSED_DIR / "05_category_inflows.csv")
category_sum = cat_df.groupby("category")["net_inflow_crore"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
category_sum.plot(kind="bar", color="orange")
plt.title("Total Net Inflows by Fund Category")
plt.xlabel("Category")
plt.ylabel("Net Inflow (₹ Crore)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Chart 5 — Top 10 funds by 3 year returns
top_funds = performance.sort_values("return_3yr_pct", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_funds, x="return_3yr_pct", y="scheme_name", hue="scheme_name", legend=False)
plt.title("Top 10 Funds by 3 Year Returns")
plt.xlabel("3 Year Return (%)")
plt.ylabel("Scheme Name")
plt.tight_layout()
plt.show()
```
