```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path().resolve()
RAW_DIR = BASE_DIR / "data" / "raw"

fund_master = pd.read_csv(RAW_DIR / "01_fund_master.csv")

# Task 6 — Explore fund master
print("Unique Fund Houses:")
print(fund_master["fund_house"].unique())

print("\nUnique Categories:")
print(fund_master["category"].unique())

print("\nUnique Sub Categories:")
print(fund_master["sub_category"].unique())

print("\nUnique Risk Grades:")
print(fund_master["risk_category"].unique())
```

```python
# Task 7 — Validate AMFI codes
nav_history = pd.read_csv(RAW_DIR / "02_nav_history.csv")

fund_master_codes = set(fund_master["amfi_code"].unique())
nav_history_codes = set(nav_history["amfi_code"].unique())

# Check if all fund_master codes exist in nav_history
missing_codes = fund_master_codes - nav_history_codes

print("Total codes in fund_master:", len(fund_master_codes))
print("Total codes in nav_history:", len(nav_history_codes))
print("Missing codes:", missing_codes if missing_codes else "None — all codes match! ✅")

print("\n--- Data Quality Summary ---")
print(f"fund_master has {len(fund_master)} schemes")
print(f"nav_history has {nav_history['amfi_code'].nunique()} unique schemes")
print(f"nav_history has {len(nav_history)} total rows")
print(f"All AMFI codes validated: {'✅ Yes' if not missing_codes else '❌ Some missing'}")
```
