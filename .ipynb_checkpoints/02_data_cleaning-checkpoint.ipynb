```python
import pandas as pd
from pathlib import Path

# Paths
BASE_DIR = Path().resolve()
RAW_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DIR = BASE_DIR / "data" / "processed"

# Load all files
fund_master = pd.read_csv(RAW_DIR / "01_fund_master.csv")
nav_history = pd.read_csv(RAW_DIR / "02_nav_history.csv")
aum = pd.read_csv(RAW_DIR / "03_aum_by_fund_house.csv")
sip = pd.read_csv(RAW_DIR / "04_monthly_sip_inflows.csv")
category = pd.read_csv(RAW_DIR / "05_category_inflows.csv")
folio = pd.read_csv(RAW_DIR / "06_industry_folio_count.csv")
performance = pd.read_csv(RAW_DIR / "07_scheme_performance.csv")
transactions = pd.read_csv(RAW_DIR / "08_investor_transactions.csv")
holdings = pd.read_csv(RAW_DIR / "09_portfolio_holdings.csv")
benchmark = pd.read_csv(RAW_DIR / "10_benchmark_indices.csv")

print("✅ All files loaded!")
```

```python
# Check for missing values in each file
dataframes = {
    "fund_master": fund_master,
    "nav_history": nav_history,
    "aum": aum,
    "sip": sip,
    "category": category,
    "folio": folio,
    "performance": performance,
    "transactions": transactions,
    "holdings": holdings,
    "benchmark": benchmark,
}

for name, df in dataframes.items():
    missing = df.isnull().sum().sum()
    print(f"{name}: {missing} missing values")
```

```python
# Check which columns have missing values in sip
print(sip.isnull().sum())
```

```python
# Fix missing values — fill yoy_growth_pct with 0 for first 12 months
sip["yoy_growth_pct"] = sip["yoy_growth_pct"].fillna(0)

# Verify it's fixed
print(sip.isnull().sum())
print("\n✅ Missing values fixed!")
```

```python
# Convert date columns to proper date format
nav_history["date"] = pd.to_datetime(nav_history["date"])
benchmark["date"] = pd.to_datetime(benchmark["date"])
transactions["transaction_date"] = pd.to_datetime(transactions["transaction_date"])
holdings["portfolio_date"] = pd.to_datetime(holdings["portfolio_date"])
aum["date"] = pd.to_datetime(aum["date"])
fund_master["launch_date"] = pd.to_datetime(fund_master["launch_date"])

# Verify
print("nav_history date type:", nav_history["date"].dtype)
print("benchmark date type:", benchmark["date"].dtype)
print("transactions date type:", transactions["transaction_date"].dtype)
print("\n✅ All dates converted!")
```

```python
# Check for duplicate rows
for name, df in dataframes.items():
    dupes = df.duplicated().sum()
    print(f"{name}: {dupes} duplicate rows")
```

```python
# Save all cleaned files to processed folder
fund_master.to_csv(PROCESSED_DIR / "01_fund_master.csv", index=False)
nav_history.to_csv(PROCESSED_DIR / "02_nav_history.csv", index=False)
aum.to_csv(PROCESSED_DIR / "03_aum_by_fund_house.csv", index=False)
sip.to_csv(PROCESSED_DIR / "04_monthly_sip_inflows.csv", index=False)
category.to_csv(PROCESSED_DIR / "05_category_inflows.csv", index=False)
folio.to_csv(PROCESSED_DIR / "06_industry_folio_count.csv", index=False)
performance.to_csv(PROCESSED_DIR / "07_scheme_performance.csv", index=False)
transactions.to_csv(PROCESSED_DIR / "08_investor_transactions.csv", index=False)
holdings.to_csv(PROCESSED_DIR / "09_portfolio_holdings.csv", index=False)
benchmark.to_csv(PROCESSED_DIR / "10_benchmark_indices.csv", index=False)

print("✅ All cleaned files saved to processed folder!")
```
